The elements of the Pauli group have another important property that relates to quantum states, which we will study now. In the next section, we will use the notions introduced here to construct stabilizer groups.

We will need to recall the eigenvectors and eigenvalues of $X, Y, Z$ are
\begin{equation*}
\begin{aligned}[c]
X\ket{+} &= \ket{+},\\
Y\ket{+\iu} &= \ket{+\iu}\\
Z\ket{0} &= \ket{0},
\end{aligned}
\qquad\qquad\qquad
\begin{aligned}[c]
X\ket{-} &= -\ket{-},\\
Y\ket{-\iu} &= -\ket{-\iu},\\
Z\ket{1} &= -\ket{1}.
\end{aligned}
\end{equation*}

## Stabilizer states
In the equations above, the ones on the left are +1 eigenstates. This means acting on the state by the corresponding operator leaves the state unchanged (even accounting for a phase). We are going to give a special name to such states.

> Given $P \in \group{P}_n$, a state $\ket{\psi}$ such that $P\ket{\psi} = \ket{\psi}$, is called a **stabilizer state** of $P$.

Let's look at some examples for $n>1$. 

**Example:** Consider, the Pauli operator $X \otimes Z$. From the equations above, it is quite easy to see that
\begin{equation}
 (X\otimes Z)\ket{+}\otimes \ket{0} = \ket{+} \otimes \ket{0}. 
\end{equation}
The tensor product of the $+1$ eigenvectors of the individual operators gives a $+1$ eigenvector of the tensor-product operator. But, wait, there is more. 
\begin{equation}
 (X\otimes Z)\ket{-}\otimes \ket{1} = (-1)^2\ket{-} \otimes \ket{1} = \ket{-} \otimes \ket{1}. 
\end{equation}
In this way, we can see that the operator $X \otimes Z$ has two $+1$ eigenvectors (it also has two $-1$ eigenvectors; why?).

Of course, any state that is a linear combination of $\ket{+} \otimes \ket{0}$ and $\ket{-} \otimes \ket{1}$ is also a $+1$ eigenstate of $X \otimes Z$,
\begin{align}
 (X\otimes Z)(\alpha\ket{+}\otimes \ket{0} + \beta \ket{-}\otimes \ket{1}) &= \alpha (X\otimes Z) \ket{+}\otimes \ket{0} + \beta (X\otimes Z) \ket{-}\otimes \ket{1}, \\
 &= \alpha\ket{+}\otimes \ket{0} + \beta \ket{-}\otimes \ket{1}.
\end{align}

Similarly, for any Pauli operator $P$ in $\group{P}_n$, we can construct states that are the $+1$ eigenvectors of the operator. 

**Example:** Here is another example,
\begin{align}
 (X \otimes I \otimes X)\ket{+}\otimes \ket{0} \otimes \ket{+} = \ket{+} \otimes \ket{0} \otimes \ket{+}, \\
 (X \otimes I \otimes X)\ket{-}\otimes \ket{0} \otimes \ket{-} = \ket{-} \otimes \ket{0} \otimes \ket{-}.
\end{align}
$I$ is special because every state is a $+1$ eigenvector of $I$. So, we could also have written any state in middle, such as
\begin{align}
 (X \otimes I \otimes X)\ket{+}\otimes \ket{-} \otimes \ket{+} = \ket{+} \otimes \ket{-} \otimes \ket{+}, \\
 (X \otimes I \otimes X)\ket{-}\otimes \ket{-} \otimes \ket{-} = \ket{-} \otimes \ket{-} \otimes \ket{-}.
\end{align}
etc. From this we infer that $X \otimes I \otimes X$ stabilizes all states that are a linear combination of 
\begin{equation}
\set{\ket{+} \otimes \ket{0} \otimes \ket{+}, \ket{-} \otimes \ket{0} \otimes \ket{-}, \ket{+} \otimes \ket{1} \otimes \ket{+}, \ket{-} \otimes \ket{1} \otimes \ket{-}}.
\end{equation}


#### Task 1 (On paper)
Determine the basis of the stabilizer states of the following operators.

* $X \otimes X \otimes X$

**SOLVE:** $X$ stabilizes the $\ket{+}$ state only. So, combinations of $\ket{-}$ state with $\ket{+}$ state should be the stabilizer states for
$X \otimes X \otimes X$. So, there are total 4 states.
  
\begin{align}
 (X \otimes X \otimes X)\ket{+}\otimes \ket{+} \otimes \ket{+} = \ket{+} \otimes \ket{+} \otimes \ket{+}, \\
 (X \otimes X \otimes X)\ket{+}\otimes \ket{-} \otimes \ket{-} = \ket{+} \otimes \ket{-} \otimes \ket{-}, \\
 (X \otimes X \otimes X)\ket{-}\otimes \ket{-} \otimes \ket{+} = \ket{-} \otimes \ket{-} \otimes \ket{+}, \\
 (X \otimes X \otimes X)\ket{-}\otimes \ket{+} \otimes \ket{-} = \ket{-} \otimes \ket{+} \otimes \ket{-}.
\end{align}

* $Z \otimes Z \otimes Z$

**SOLVE:** $Z$ stabilizes the $\ket{0}$ state only. So, combinations of $\ket{0}$ state with $\ket{1}$ state should be the stabilizer states for $Z \otimes Z \otimes Z$. So, there are total 4 states.
  
\begin{align}
 (Z \otimes Z \otimes Z)\ket{0}\otimes \ket{0} \otimes \ket{0} = \ket{0} \otimes \ket{0} \otimes \ket{0}, \\
 (Z \otimes Z \otimes Z)\ket{0}\otimes \ket{1} \otimes \ket{1} = \ket{0} \otimes \ket{1} \otimes \ket{1}, \\
 (Z \otimes Z \otimes Z)\ket{1}\otimes \ket{1} \otimes \ket{0} = \ket{1}\otimes \ket{1} \otimes \ket{0}, \\
 (Z \otimes Z \otimes Z)\ket{1}\otimes \ket{0} \otimes \ket{1} = \ket{1}\otimes \ket{0} \otimes \ket{1}.
\end{align}

* $Z \otimes I \otimes X$

**SOLVE:** $Z$ stabilizes the $\ket{0}$ state only. $X$ stabilizes the $\ket{+}$ state only. So, combinations of $\ket{0}$, $\ket{1}$ states with $\ket{+}$ and $\ket{-}$ states should be the stabilizer states for $Z \otimes I \otimes X$. So, there are total 8 states. On the $+$,$-$ basis  it has 4 states. And on the $0$,$1$ basis there are other 4 states.
  
\begin{align}
 (Z \otimes I \otimes X)\ket{0}\otimes \ket{0} \otimes \ket{+} = \ket{0} \otimes \ket{0} \otimes \ket{+}, \\
 (Z \otimes I \otimes X)\ket{1}\otimes \ket{0} \otimes \ket{-} = \ket{1}\otimes \ket{0} \otimes \ket{-}, \\
 (Z \otimes I \otimes X)\ket{0}\otimes \ket{1} \otimes \ket{+} = \ket{0}\otimes \ket{1} \otimes \ket{+}, \\
 (Z \otimes I \otimes X)\ket{1}\otimes \ket{1} \otimes \ket{-} = \ket{1}\otimes \ket{1} \otimes \ket{-}, \\
  (Z \otimes I \otimes X)\ket{0}\otimes \ket{+} \otimes \ket{+} = \ket{0} \otimes \ket{+} \otimes \ket{+}, \\
  (Z \otimes I \otimes X)\ket{1}\otimes \ket{+} \otimes \ket{-} = \ket{1}\otimes \ket{+} \otimes \ket{-}, \\
  (Z \otimes I \otimes X)\ket{0}\otimes \ket{-} \otimes \ket{+} = \ket{0}\otimes \ket{-} \otimes \ket{+}, \\
  (Z \otimes I \otimes X)\ket{1}\otimes \ket{-} \otimes \ket{-} = \ket{1}\otimes \ket{-} \otimes \ket{-}.
\end{align}

* $Z \otimes X \otimes X \otimes Z$

**SOLVE:** $Z$ stabilizes the $\ket{0}$ state only. $X$ stabilizes the $\ket{+}$ state only. So, combinations of $\ket{0}$, $\ket{1}$ states with $\ket{+}$ and $\ket{-}$ states should be the stabilizer states for $Z \otimes I \otimes X$. So, there are total 8 states. On the $+$,$-$ basis  it has 4 states. And on the $0$,$1$ basis there are other 4 states.
  
\begin{align}
 (Z \otimes X \otimes X \otimes Z)\ket{0}\otimes \ket{+} \otimes \ket{+} \otimes \ket{0}= \ket{0}\otimes \ket{+} \otimes \ket{+} \otimes \ket{0}, \\
 (Z \otimes X \otimes X \otimes Z)\ket{1}\otimes \ket{+} \otimes \ket{+} \otimes \ket{1} = \ket{1}\otimes \ket{+} \otimes \ket{+} \otimes \ket{1}, \\
 (Z \otimes X \otimes X \otimes Z)\ket{0}\otimes \ket{-} \otimes \ket{-} \otimes \ket{0} = \ket{0}\otimes \ket{-} \otimes \ket{-} \otimes \ket{0}, \\
 (Z \otimes X \otimes X \otimes Z)\ket{1}\otimes \ket{-} \otimes \ket{-} \otimes \ket{1} = \ket{1}\otimes \ket{-} \otimes \ket{-} \otimes \ket{1}, \\
 (Z \otimes X \otimes X \otimes Z)\ket{0}\otimes \ket{+} \otimes \ket{-} \otimes \ket{1}= \ket{0}\otimes \ket{+} \otimes \ket{-} \otimes \ket{1}, \\
 (Z \otimes X \otimes X \otimes Z)\ket{1}\otimes \ket{-} \otimes \ket{+} \otimes \ket{0}= \ket{1}\otimes \ket{-} \otimes \ket{+} \otimes \ket{0}, \\
 (Z \otimes X \otimes X \otimes Z)\ket{1}\otimes \ket{+} \otimes \ket{-} \otimes \ket{0}= \ket{1}\otimes \ket{+} \otimes \ket{-} \otimes \ket{0}.
\end{align}

Do you see a pattern in the number of states and the size of the operator?

## Stabilizer states of multiple operators
We will now show that it is possible to construct states that are the simultaneously stabilized by multiple Pauli operators. Let's start with some examples.

**Example:** Consider, the operators $P_1 = Z \otimes Z \otimes I$ and $P_2= Z \otimes I \otimes Z$, both in $\group{P}_3$.

* $Z \otimes Z \otimes I$ stabilizes four orthogonal states $\set{\ket{000}, \ket{001}, \ket{110}, \ket{111}}$.
* $Z \otimes I \otimes Z$ stabilizes four orthogonal states $\set{\ket{000}, \ket{010}, \ket{101}, \ket{111}}$.

The intersection of these sets is $\set{\ket{000}, \ket{111}}$. This means that any state that is a linear combination of these two states will be stabilized by both $P_1$ and $P_2$.

Remember, how $\set{\ket{000}, \ket{111}}$ were the basis states of the quantum repetition code for bit-flip codes. It turns out that code is a stabilizer code. We will soon show this, if you have some patience.

#### Task 2 (On paper)
Determine states that are simultaneously stabilized by $P_1 = X \otimes X \otimes I = X_0X_1$ and $P_2 = X \otimes I \otimes X = X_0X_2$. (The subscript based notation will ensure you don't injure your wrist!)

**SOLVE:** So, $P_1 = X \otimes X \otimes I = X_0X_1$ stabilizes $\set{\ket{+++}, \ket{--+}, \ket{++-}, \ket{---}}$.  While, $P_2 = X \otimes I \otimes X = X_0X_2$ stabilizes $\set{\ket{+++}, \ket{+-+}, \ket{-+-}, \ket{---}}$. So, simultaneously stabilized states are $\ket{+++}$ and $\ket{---}$. 

Have you seen them before? 

**SOLVE:** Yes. In Phase flip error correction code. 

#### Task 3 (On paper)
It is not necessary that two operators have simultaneously stabilized states. 

* Do $P_1 = Z_0Z_1$ and $P_2=X_0X_1$ have any simultaneously stabilized states?

**SOLVE:**  $P_1 = Z_0Z_1$ stabilizes four orthogonal states $\set{\ket{000}, \ket{001}, \ket{110}, \ket{111}}$. And $P_2=X_0X_1$ stabilizes $\set{\ket{+++}, \ket{--+}, \ket{++-}, \ket{---}}$. So, they $P_1$ and $P_2$ have no simultaneously stabilized basis states. So, they don't stabilize any basis states. But they can stabilize linear combinations of the basis states. For example, $P_1 = Z_0Z_1$ and $P_2=X_0X_1$ stabilize the bell pair state: $\frac{\ket{00} + \ket{11}}{\sqrt{2}}$.

* What about $P_1=Z_0Z_1$ and $P_2=X_0X_2$?

**SOLVE:**  $P_1 = Z_0Z_2$ four orthogonal states $\set{\ket{000}, \ket{010}, \ket{101}, \ket{111}}$. And $P_2=X_0X_2$ stabilizes $\set{\ket{+++}, \ket{+-+}, \ket{-+-}, \ket{---}}$. So, they $P_1$ and $P_2$ have no simultaneously stabilized states. So, they don't stabilize any basis states. Also, they don't stabilize any linear combinations simultaneously. 

Becuase, the stabilized states of $P_2$ are:-
$$
 X_0X_2 \frac{\ket{010} + \ket{111}}{2\sqrt{2}} = \frac{\ket{010} + \ket{111}}{2\sqrt{2}}
$$
$$
 X_0X_2 \frac{\ket{000} + \ket{101}}{2\sqrt{2}} = \frac{\ket{000} + \ket{101}}{2\sqrt{2}}

$$
 X_0X_2 \frac{\ket{011} + \ket{110}}{2\sqrt{2}} = \frac{\ket{011} + \ket{110}}{2\sqrt{2}}

$$
 X_0X_2 \frac{\ket{001} + \ket{100}}{2\sqrt{2}} = \frac{\ket{001} + \ket{100}}{2\sqrt{2}}.

while, $P_1$ stabilizes,
$$
 Z_0X_1 \frac{\ket{000} + \ket{111}}{2\sqrt{2}} = \frac{\ket{010} + \ket{111}}{2\sqrt{2}}
